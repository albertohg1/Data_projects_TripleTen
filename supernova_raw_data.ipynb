```python
file_path='/content/drive/MyDrive/TripleTen/Externship -- Supernova/raw_data.csv'
```

```python
import pandas as pd
df = pd.read_csv(file_path)
df.head()
```

```python
df['insured_address_state'] = df['insured_address_state'].replace('Québec', 'Quebec')
```

```python
df = df[df['Industry Code'].between(1111, 9281)]
```

```python
df = df.rename(columns={'Industry Code': 'industry_code'})
df = df[df['industry_code'].between(1111, 9281)]
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
print("Before:", df.shape)
df = df.drop_duplicates()
print("After:", df.shape)
```

```python
# Clean and standardize Category names
category_clean_map = {
    'Health Serv.': 'Health Services',
    'B&P Services': 'Business & Personal Services',
    'Hospitality': 'Hospitality',
    'Retail': 'Retail',
    'Wholesale': 'Wholesale'
}

df['Category'] = df['Category'].str.strip().map(category_clean_map)
```

```python
df['Category'].value_counts()
```

```python
potential_duplicates = df[df.duplicated(subset=[
    'industry_code', 'insured_address_state', 'Category',
    'input_gl_revenue', 'input_gl_employee', 'input_gl_gl_limit',
    'input_property_bldg_limit', 'input_property_content_limit'
], keep=False)]  # keep=False shows all duplicate pairs

# See how many we found:
print("Potential duplicates found:", potential_duplicates.shape[0])
```

```python
potential_duplicates.head()
```

```python
output_path = '/content/drive/My Drive/TripleTen/Externship -- Supernova/cleaned_supernova_data.csv'
df.to_csv(output_path, index=False)

print("✅ Cleaned dataset saved successfully!")
```
